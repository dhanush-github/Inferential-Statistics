```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set a seed for reproducibility
np.random.seed(43)

# Number of random samples to generate
sample_size = 1000
```

```python
# Normal Distribution
def generate_normal(mu, sigma, size):
    u1 = np.random.rand(size)
    u2 = np.random.rand(size)
    z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    return mu + sigma * z0

data_normal = generate_normal(mu=0, sigma=1, size=sample_size)
```

```python
# Uniform Distribution
def generate_uniform(a, b, size):
    return a + (b - a) * np.random.rand(size)

data_uniform = generate_uniform(a=0, b=1, size=sample_size)
```

```python
# Poisson Distribution
def generate_poisson(mu, size):
    L = np.exp(-mu)
    k = 0
    p = 1
    while p > L:
        k += 1
        p *= np.random.rand()
    return k - 1

data_poisson = np.array([generate_poisson(mu=3, size=1) for _ in range(sample_size)])
```

```python
# Exponential Distribution
def generate_exponential(scale, size):
    u = np.random.rand(size)
    return -scale * np.log(1 - u)

data_exponential = generate_exponential(scale=2, size=sample_size)
```

```python
# Plotting histograms
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(data_normal, bins=30, color='blue', alpha=0.7)
plt.title('Normal Distribution')

plt.subplot(2, 2, 2)
plt.hist(data_uniform, bins=30, color='green', alpha=0.7)
plt.title('Uniform Distribution')

plt.subplot(2, 2, 3)
plt.hist(data_poisson, bins=30, color='orange', alpha=0.7)
plt.title('Poisson Distribution')

plt.subplot(2, 2, 4)
plt.hist(data_exponential, bins=30, color='red', alpha=0.7)
plt.title('Exponential Distribution')



plt.tight_layout()
plt.show()
```
