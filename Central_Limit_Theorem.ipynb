## Central Limit Theorem

The Central Limit Theorem (CLT) is a fundamental statistical principle that elucidates the asymptotic behavior of the distribution of the sample mean of a sufficiently large number of independent and identically distributed random variables. It asserts that, irrespective of the underlying distribution of the individual variables, the sum (or average) of these variables converges to a normal (Gaussian) distribution as the sample size increases.

The prerequisites for the Central Limit Theorem are threefold:

1. **Independence:** The random variables must be mutually independent.
2. **Identical Distribution:** The random variables must be drawn from an identical probability distribution.
3. **Adequate Sample Size:** Although the notion of "adequate" lacks strict universal definition, a common heuristic stipulates that a sample size exceeding 30 is generally sufficient for the CLT to manifest its convergence properties.

The Central Limit Theorem holds profound significance in statistical inference, furnishing a theoretical foundation for the application of normal distribution-based methodologies in situations where the original population distribution is not normal. Its ubiquity in statistical theory and practice renders the CLT a cornerstone for various statistical analyses, hypothesis testing, and the formulation of confidence intervals.

### Central Limit Theorem with a BiModal Distribution

```python
a = np.random.normal(100, 5, 1000)
```

```python
b = np.random.normal(80, 5, 1000)
```

```python
bm_data = np.append(a, b)
np.random.shuffle(bm_data)
```

```python
sns.displot(bm_data)
plt.xlim(60,120)
```

```python
bm_2 = bm_data.reshape(1000, 2).mean(axis=1)
sns.displot(bm_2)
plt.xlim(60,120)
```

```python
bm_4 = bm_data.reshape(500, 4).mean(axis=1)
sns.displot(bm_4)
plt.xlim(60,120)
```

```python
bm_10 = bm_data.reshape(200, 10).mean(axis=1)
sns.displot(bm_10)
plt.xlim(60,120)
```
